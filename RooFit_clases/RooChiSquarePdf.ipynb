```python
import ROOT
from ROOT import RooFit, RooRealVar, RooChiSquarePdf
from IPython.display import Image, display

# Define the variable x
x = ROOT.RooRealVar("x", "x", 0, 10)  # Range for the variable x

# Define the degrees of freedom for the Chi-squared distribution
df = ROOT.RooRealVar("df", "degrees of freedom", 3, 1, 10)  # Degrees of freedom

# Create a Chi-squared PDF
chi_square_pdf = RooChiSquarePdf("chi_square_pdf", "Chi-squared PDF", x, df)

# Generate toy data from the Chi-squared PDF
data = chi_square_pdf.generate(ROOT.RooArgSet(x), 10000)

# Create a frame for plotting
xframe = x.frame(RooFit.Title("Chi-squared PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Chi-squared PDF on the same frame
chi_square_pdf.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Chi-squared PDF", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("ChiSquarePdfPlot.png")

# Display the image in Jupyter notebook
display(Image(filename="ChiSquarePdfPlot.png"))
```
