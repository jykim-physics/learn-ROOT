```python
import ROOT
from ROOT import RooFit, RooRealVar, RooChebychev
from IPython.display import Image, display

# Define the variable x
x = ROOT.RooRealVar("x", "x", -5, 5)  # Range for the variable x

# Define coefficients for the Chebyshev polynomial
c0 = ROOT.RooRealVar("c0", "c0", 0.5, -1, 1)  # Coefficient for the constant term
c1 = ROOT.RooRealVar("c1", "c1", 0.3, -1, 1)  # Coefficient for the linear term
c2 = ROOT.RooRealVar("c2", "c2", -0.1, -1, 1) # Coefficient for the quadratic term

# Create a Chebyshev polynomial PDF
cheby = RooChebychev("cheby", "Chebyshev PDF", x, ROOT.RooArgList(c0, c1, c2))

# Generate toy data from the Chebyshev PDF
data = cheby.generate(ROOT.RooArgSet(x), 10000)

# Create a frame for plotting
xframe = x.frame(RooFit.Title("Chebyshev Polynomial PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Chebyshev polynomial on the same frame
cheby.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Chebyshev Polynomial PDF", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("ChebyshevPlot.png")

# Display the image in Jupyter notebook
display(Image(filename="ChebyshevPlot.png"))
```
