```python
import ROOT
from ROOT import RooFit, RooRealVar, RooBreitWigner
from IPython.display import Image, display

# Define the mass variable (x)
x = ROOT.RooRealVar("x", "mass", 0, 10)  # Mass variable range

# Define the parameters for the Breit-Wigner PDF
mean = ROOT.RooRealVar("mean", "mean mass", 5.0, 0, 10)  # Resonance mean mass
width = ROOT.RooRealVar("width", "width", 1.0, 0.1, 5.0)  # Width of the resonance

# Create the Breit-Wigner PDF
bw_pdf = ROOT.RooBreitWigner("bw_pdf", "Breit-Wigner PDF", x, mean, width)

# Generate toy data from the Breit-Wigner PDF
data = bw_pdf.generate(ROOT.RooArgSet(x), 10000)

# Create a frame for plotting
xframe = x.frame(RooFit.Title("Breit-Wigner PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Breit-Wigner PDF on the same frame
bw_pdf.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Breit-Wigner", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("BreitWigner_plot.png")

# Display the image in Jupyter notebook
display(Image(filename="BreitWigner_plot.png"))
```
