```python
import ROOT
from ROOT import RooFit, RooRealVar, RooCBShape
from IPython.display import Image, display

# Define the variable x
x = ROOT.RooRealVar("x", "x", -5, 10)  # Range for the variable x

# Define parameters for the Crystal Ball function
mean = ROOT.RooRealVar("mean", "mean", 2.0, -5, 10)       # Mean of the distribution
sigma = ROOT.RooRealVar("sigma", "sigma", 1.0, 0.1, 5)    # Width of the Gaussian core
alpha = ROOT.RooRealVar("alpha", "alpha", 1.5, 0, 3)      # Alpha parameter for the tail
n = ROOT.RooRealVar("n", "n", 2.0, 1, 5)                  # Power of the tail

# Create a Crystal Ball shape PDF
cb_shape = RooCBShape("cb_shape", "Crystal Ball PDF", x, mean, sigma, alpha, n)

# Generate toy data from the Crystal Ball PDF
data = cb_shape.generate(ROOT.RooArgSet(x), 10000)

# Create a frame for plotting
xframe = x.frame(RooFit.Title("Crystal Ball Shape PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Crystal Ball shape on the same frame
cb_shape.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Crystal Ball Shape PDF", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("CrystalBallShape_plot.png")

# Display the image in Jupyter notebook
display(Image(filename="CrystalBallShape_plot.png"))
```
