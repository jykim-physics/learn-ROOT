```python
import ROOT
from ROOT import RooFit, RooRealVar, RooBernstein, RooArgSet
from IPython.display import Image, display

# Define the observable x
x = RooRealVar("x", "x", 0, 10)

# Define the coefficients for the Bernstein polynomial
c0 = RooRealVar("c0", "coefficient 0", 0.2, 0.0, 1.0)
c1 = RooRealVar("c1", "coefficient 1", 0.3, 0.0, 1.0)
c2 = RooRealVar("c2", "coefficient 2", 0.5, 0.0, 1.0)

# Create a Bernstein PDF with 3 coefficients
bernstein = RooBernstein("bernstein", "Bernstein PDF", x, RooArgSet(c0, c1, c2))

# Generate a toy dataset from the Bernstein PDF
data = bernstein.generate(RooArgSet(x), 10000)

# Create a frame to plot x
xframe = x.frame(RooFit.Title("Bernstein Polynomial PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Bernstein PDF on the same frame
bernstein.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Bernstein Polynomial", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("Bernstein_PDF_plot.png")

# Display the image in Jupyter notebook
display(Image(filename="Bernstein_PDF_plot.png"))
```
