```python
import ROOT
from ROOT import RooFit, RooRealVar, RooDecay, RooArgSet, RooGaussModel
from IPython.display import Image, display

# Define the variable t (time)
t = ROOT.RooRealVar("t", "Time", 0, 10)  # Time range from 0 to 10

# Define the lifetime (tau) and decay constant (gamma)
tau = ROOT.RooRealVar("tau", "Lifetime", 1.0, 0.1, 10)  # Lifetime
gamma = ROOT.RooRealVar("gamma", "Decay Constant", 1/tau.getVal())  # Decay constant

# Define the Gaussian resolution model for time resolution
timeResMean = RooRealVar("timeResMean", "Time Resolution Mean", 0.0)
timeResSigma = RooRealVar("timeResSigma", "Time Resolution Sigma", 0.1)
resModel = RooGaussModel("resModel", "Gaussian Time Resolution", t, timeResMean, timeResSigma)


# Create a decay PDF using RooDecay
decay_pdf = RooDecay("decay_pdf", "Decay PDF", t, tau,resModel, ROOT.RooDecay.SingleSided)

# Generate toy data from the decay PDF
data = decay_pdf.generate(ROOT.RooArgSet(t), 10000)

# Create a frame for plotting
tframe = t.frame(RooFit.Title("Decay PDF"))

# Plot the generated data
data.plotOn(tframe)

# Plot the decay PDF on the same frame
decay_pdf.plotOn(tframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Decay PDF", 800, 600)
tframe.Draw()

# Save the plot as an image
c1.SaveAs("DecayPdfPlot.png")

# Display the image in Jupyter notebook
display(Image(filename="DecayPdfPlot.png"))
```
