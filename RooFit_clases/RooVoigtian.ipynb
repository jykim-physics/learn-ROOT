```python
import ROOT

# Define a RooRealVar, the variable on which the PDF depends
x = ROOT.RooRealVar("x", "x", -10, 10)

# Define the parameters of the Gaussian: mean and sigma
mean = ROOT.RooRealVar("mean", "mean of gaussian", 0, -10, 10)
sigma = ROOT.RooRealVar("sigma", "width of gaussian", 2, 0.1, 10)

# Create the Gaussian PDF
gauss = ROOT.RooGaussian("gauss", "gaussian PDF", x, mean, sigma)

# Generate a toy dataset with 1000 events
data = gauss.generate(ROOT.RooArgSet(x), 1000)

# Plot the data and PDF on a frame
xframe = x.frame(ROOT.RooFit.Title("Gaussian PDF and generated data"))
data.plotOn(xframe)
gauss.plotOn(xframe)

# Draw the frame on a canvas
c = ROOT.TCanvas("c", "Gaussian PDF", 800, 600)
xframe.Draw()
# c.SaveAs("gaussian_fit.pdf")
c.SaveAs("gaussian_fit.png")

from IPython.display import Image
Image("gaussian_fit.png")
```

```python
import ROOT

# Define the variable
x = ROOT.RooRealVar("x", "x", -10, 10)

# Voigtian PDF parameters
mean = ROOT.RooRealVar("mean", "mean", 0, -10, 10)
width = ROOT.RooRealVar("width", "width of Lorentzian", 1, 0.1, 5)
sigma = ROOT.RooRealVar("sigma", "width of Gaussian", 1, 0.1, 5)

# Create the Voigtian PDF
voigt = ROOT.RooVoigtian("voigt", "Voigtian PDF", x, mean, width, sigma)

# Generate toy data
data = voigt.generate(ROOT.RooArgSet(x), 1000)

# Fit the PDF to the data
voigt.fitTo(data)

# Plot the results
xframe = x.frame(ROOT.RooFit.Title("Voigtian PDF"))
data.plotOn(xframe)
voigt.plotOn(xframe)

# Draw the frame
c = ROOT.TCanvas("c", "Voigtian Fit", 800, 600)
xframe.Draw()
c.SaveAs("voigtian_fit.png")

from IPython.display import Image
Image("voigtian_fit.png")
```
