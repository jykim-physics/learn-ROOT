```python
import ROOT
from ROOT import RooFit
from IPython.display import Image, display

# Define the two variables
x = ROOT.RooRealVar("x", "x", -10, 10)
y = ROOT.RooRealVar("y", "y", -10, 10)

# Generate a 2D dataset based on a Gaussian distribution for both x and y
mean_x = ROOT.RooRealVar("mean_x", "mean of x", 0)
sigma_x = ROOT.RooRealVar("sigma_x", "sigma of x", 2)

mean_y = ROOT.RooRealVar("mean_y", "mean of y", 0)
sigma_y = ROOT.RooRealVar("sigma_y", "sigma of y", 2)

gauss_x = ROOT.RooGaussian("gauss_x", "Gaussian for x", x, mean_x, sigma_x)
gauss_y = ROOT.RooGaussian("gauss_y", "Gaussian for y", y, mean_y, sigma_y)

# Create a RooProdPdf (product of the two PDFs, to simulate 2D Gaussian)
model = ROOT.RooProdPdf("model", "gauss_x * gauss_y", ROOT.RooArgList(gauss_x, gauss_y))

# Generate a toy dataset
data = model.generate(ROOT.RooArgSet(x, y), 10000)

# Create a 2D keys PDF from the data
keys_pdf = ROOT.RooNDKeysPdf("keys_pdf", "2D KDE", ROOT.RooArgList(x, y), data, "a", 1.0)

# Plot the resulting 2D PDF
xframe = x.frame(RooFit.Title("Projection of KDE on x"))
keys_pdf.plotOn(xframe)

yframe = y.frame(RooFit.Title("Projection of KDE on y"))
keys_pdf.plotOn(yframe)

# Create a 2D plot using a histogram for better visualization
hist = keys_pdf.createHistogram("hist2D", x, ROOT.RooFit.Binning(50), ROOT.RooFit.YVar(y, ROOT.RooFit.Binning(50)))

# Drawing and saving the images
c1 = ROOT.TCanvas("c1", "2D KDE - x projection", 800, 600)
xframe.Draw()
c1.SaveAs("2DKeysPDF_x_projection.png")  # Save as PNG for display in Jupyter notebook

c2 = ROOT.TCanvas("c2", "2D KDE - y projection", 800, 600)
yframe.Draw()
c2.SaveAs("2DKeysPDF_y_projection.png")  # Save as PNG for display

c3 = ROOT.TCanvas("c3", "2D KDE - 2D Histogram", 800, 600)
hist.Draw("COLZ")
c3.SaveAs("2DKeysPDF_2D_histogram.png")  # Save as PNG for display

# Display the images in Jupyter notebook
display(Image(filename="2DKeysPDF_x_projection.png"))
display(Image(filename="2DKeysPDF_y_projection.png"))
display(Image(filename="2DKeysPDF_2D_histogram.png"))
```
