```python
import ROOT
from ROOT import RooFit
from IPython.display import Image, display

# Define the variable x
x = ROOT.RooRealVar("x", "x", 0, 10)

# Define parameters for RooArgusBG
cutoff = ROOT.RooRealVar("cutoff", "cutoff", 8, 0, 10)  # Cutoff point
c = ROOT.RooRealVar("c", "c", -1, -10, 0)               # Shape parameter

# Create an Argus background PDF
argus = ROOT.RooArgusBG("argus", "Argus PDF", x, cutoff, c)

# Generate a toy dataset from the Argus PDF
data = argus.generate(ROOT.RooArgSet(x), 10000)

# Create a frame for plotting x
xframe = x.frame(RooFit.Title("Argus Background PDF"))

# Plot the generated data
data.plotOn(xframe)

# Plot the Argus background PDF on the same frame
argus.plotOn(xframe)

# Create a canvas to draw the plot
c1 = ROOT.TCanvas("c1", "Argus Background", 800, 600)
xframe.Draw()

# Save the plot as an image
c1.SaveAs("ArgusBG_plot.png")

# Display the image in Jupyter notebook
display(Image(filename="ArgusBG_plot.png"))
```
